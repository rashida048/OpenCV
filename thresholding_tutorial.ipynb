```python
import cv2 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread("meter.jpg")
```

```python
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (7, 7), 0)
```

```python
(T, thresh) = cv2.threshold(blurred, 45, 255,
                              cv2.THRESH_BINARY)
plt.imshow(thresh)
```

```python
T
```

```python
cv2.imwrite('a.png', thresh)
```

```python
(T, threshInv) = cv2.threshold(blurred, 45, 255,
                              cv2.THRESH_BINARY_INV)
plt.imshow(threshInv)
```

```python
cv2.imwrite('aa.png', threshInv)
```

```python
masked = cv2.bitwise_and(image, image, mask=thresh)
plt.imshow(masked)
```

```python
(T, threshOtsu) = cv2.threshold(blurred, 0, 255, 
                cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
T
```

```python
plt.imshow(threshOtsu)
```

```python
cv2.imwrite("au.jpg", threshOtsu)
```

```python
masked = cv2.bitwise_and(image, image, mask=threshOtsu)
```

```python
(T, threshInvOtsu) = cv2.threshold(blurred, 0, 255, 
                cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
cv2.imwrite("aiu.jpg", threshInvOtsu)
```

```python
masked1 = cv2.bitwise_and(image, image, mask=threshInvOtsu)
```

```python
cv2.imwrite('amasked.jpg', masked1)
```
